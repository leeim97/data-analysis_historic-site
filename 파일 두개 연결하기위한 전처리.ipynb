#### 한 파일에는 이름에 특정부호와 공백이들어있고 나머지 파일에는 공백없는 이름이라 이름 맞춰주기

```python
names=[]
for name in df2['관광지']:
    names.append(re.sub(r'[^\uAC00-\uD7A3]','',name))
names
```

```python
df2['관광지']=names
df2.관광지
```

#### 첫번째 파일에서 이름이 같은 유적지에 두번째 파일에 위도,경도,도로명,건물번호 넣어줌

```python
for i,name2 in zip(df2.index,df2['관광지']):
    for name,la,lo,cate,road,ro_No in zip(df['POI_NM'],df['LC_LA'],df['LC_LO'],df['CL_NM'],
                                          df['RDNMADR_NM'],df['BULD_NO']):
        if name2 == name:
            df2.loc[i,'위도']=la
            df2.loc[i,'경도']=lo
            df2.loc[i,'범주']=cate
            df2.loc[i,'도로명']=road
            df2.loc[i,'건물번']=ro_No
df2
       
```

```python
df2.to_excel('주요관광지점 입장객(전국)_위도경도_추가버전.xlsx', index=False)
```
