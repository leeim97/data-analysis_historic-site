# 별점 관련 데이터처리

```python
df = pd.read_excel('주요관광지점 입장객(전국).xlsx')
```

```python
# 시도 이름목록 추출
datas=list(df['시도'].unique())
```

```python
import os
```

```python
def star(place):
    # directory변수를 보면 알수있듯이 무조건 data폴더안에 시도 폴더가 있어야됌
    sido=place
    directory = 'data/'+sido
    
    # 예시) 충청북도 안에있는 모든 파일 갖고오기
    files = os.listdir(directory)
    excel_files = []
    # 파일 목록 중에서 엑셀 파일인 것만 추려내기
    for file in files:
        if file.endswith('.xlsx') or file.endswith('.xls'):
            excel_files.append(file)
    
    
    
    
    
    # 엑셀파일당 평균 평점과 장소 이름을 모아둘 리스트
    rating=[]
    
    for excel_file in excel_files:
        # 파일 경로
        file_path = os.path.join(directory, excel_file)
        # 엑셀파일을 데이터 프레임화 시켜서 각 엑셀파일(장소)마다 평점의 평균을 뽑는다
        df = pd.read_excel(file_path)
        rating.append((df['평점'].mean(),excel_file.split('.')[0]))
    
    # 예) 충청북도의 각 장소마다의 평점을 갖고와서 내림차순으로 정렬
    rating.sort(reverse=True)
    # 예) 충청북도에서 별점 제일 높은 5개
    star_top10 = rating[0:11]
    # 예) 충청북도 전체 별점 평균
    star_sum=0
    cnt=0
    for rat in rating:
        print(rat[0])
        if rat[0] != 0:
            star_sum+=rat[0]
            cnt+=1
    
    star_mean = round(star_sum/cnt,2)
    
    return star_top10,star_mean
```

#### 각각의 장소 평균 평점,각 시도별 장소 평균평점 순위 ,전체 평균평점 높은순위 구하기

```python
total_sido=[]
total_place=[]
for place in datas:
    star_top10,star_mean=star(place)
    x = [sc[1] for sc in star_top10]
    y = [sc[0] for sc in star_top10]
    df_sido = pd.DataFrame({'name':x,'rating':y})
    df_sido.to_csv(f'data/place/{place}_별점.csv', index=False)
    total_sido.append((star_mean,place))
    total_place.append(star_top10)
```

```python
# 시도별 평점 순위
total_sido.sort(reverse=True)
x = [sc[1] for sc in total_sido]
y = [sc[0] for sc in total_sido]
df_total_sido = pd.DataFrame({'name':x,'rating':y})
df_total_sido.to_csv(f'data/place/전국_시도_별점순위.csv', index=False)
```
